#  HOW MUCH DOES A STUDENT EARNS BASED ON WHAT HE STUDIES
#### The aim of the project is to visualize how much does a student earns based on his majors during college.

1. Firstly, removing the rows having null values.

```python
import pandas as pd
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
recent_grads = pd.read_csv('recent-grads.csv') ## reading the dataset
print(recent_grads.iloc[0]) ## reading the headers of recent-grad dataset
print('----------------------------------------')
print(recent_grads.describe())
print('-----------------------------------------')
raw_data_count = recent_grads.count()
print(raw_data_count)
print('-----------------------------------------')
raw_data_count  =  len(recent_grads.index)
print(raw_data_count)
recent_grads = recent_grads.dropna(axis=0)
cleaned_data_count = recent_grads.count()
print(cleaned_data_count)
cleaned_data_count = len(recent_grads.index)
print('------------------------------------------')
print(cleaned_data_count)
```

5. Bar Plots

```python
recent_grads[:10].plot.bar(x='Major', y='ShareWomen', legend = False)
recent_grads[163:].plot.bar(x='Major', y='ShareWomen', legend=False)
```

Here, we can obeserve that more percentages of women are engaged in the jobs with low paying majors. Evidently, women are paid less. 

```python
recent_grads[:10].plot.bar(x='Major', y='Unemployment_rate', legend=False)
recent_grads[163:].plot.bar(x='Major', y='Unemployment_rate', legend=False)
```

Comparing the unemployement rates, from the charts above we can say that Unemployment in low paying is comparitively more. This can be due to less amount of students opting for majors that lead to low paying jobs.
